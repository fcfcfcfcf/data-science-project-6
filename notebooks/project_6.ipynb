```python
import pandas as pd

```

First let's read in the datasets, I've chosen to use the Twitter dataset

```python
aapl = pd.read_csv('../data/Twitter_volume_AAPL.csv')
amzn = pd.read_csv('../data/Twitter_volume_AMZN.csv')
crm = pd.read_csv('../data/Twitter_volume_CRM.csv')
cvs = pd.read_csv('../data/Twitter_volume_CVS.csv')
fb = pd.read_csv('../data/Twitter_volume_FB.csv')
goog = pd.read_csv('../data/Twitter_volume_GOOG.csv')
ibm = pd.read_csv('../data/Twitter_volume_IBM.csv')
ko = pd.read_csv('../data/Twitter_volume_KO.csv')
pfe = pd.read_csv('../data/Twitter_volume_PFE.csv')
ups = pd.read_csv('../data/Twitter_volume_UPS.csv')
```

Next we will merge the dataset into one consolidated dataframe

```python
merged_tweets = pd.DataFrame()
merged_tweets = pd.merge(aapl, amzn, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value_x': 'AAPL', 'value_y': 'AMZN'})
merged_tweets = pd.merge(merged_tweets, crm, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value': 'CRM'})
merged_tweets = pd.merge(merged_tweets, cvs, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value': 'CVS'})
merged_tweets = pd.merge(merged_tweets, fb, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value': 'FB'})
merged_tweets = pd.merge(merged_tweets, goog, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value': 'GOOG'})
merged_tweets = pd.merge(merged_tweets, ibm, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value': 'IBM'})
merged_tweets = pd.merge(merged_tweets, ko, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value': 'KO'})
merged_tweets = pd.merge(merged_tweets, pfe, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value': 'PFE'})
merged_tweets = pd.merge(merged_tweets, ups, on='timestamp')
merged_tweets = merged_tweets.rename(columns={'value': 'UPS'})

merged_tweets.plot()
```

It looks like AAPL is by far the most mentioned ticker out of the set of tickers given (this makes sense of course, it's Apple we are talking about after all).

One useful feature to add could be the total number of tweets across all tickers. This might provide insight to the reason for an increase in volume, and whether or not that is an independent event or tied to an increase in twitter activity across all tickers.

```python
merged_tweets['total_volume'] = merged_tweets.sum(axis=1)
merged_tweets
```

Let's plot AAPL against the total number of tweets for all tickers (I've plotted them seperatly because it was hard to tell what was going on when consolidating into one plot, even with seperate y-axis).

```python
import matplotlib.pyplot as plt


merged_tweets['total_volume'].plot()
plt.show()

```

```python

merged_tweets['AAPL'].plot()
```

These graphs look almost identical, but this is actually AAPL's share of the tweets. Here is UPS for reference.

```python
merged_tweets['UPS'].plot()
```

Let's attempt to detect anomolies. The first approach I'd like to try is using an Isolation Forest Algorithm. I belive that the algorithm will generally recognize the cluster of 'normal' numbers of tweets while isolating the large tweet amounts. We don't have many features, so isolation should be fairly straightforward due to low dimensionality.



```python
from sklearn.ensemble import IsolationForest

X = merged_tweets[['AAPL', 'total_volume']]
clf = IsolationForest().fit(X)
classifications = pd.DataFrame(clf.predict(X))
classifications = classifications.replace({1: 0, -1: 1})
merged_tweets['AAPL-outliers'] = classifications
merged_tweets[['AAPL', 'AAPL-outliers']].plot()

```

```python
merged_tweets[['AAPL', 'AAPL-outliers']].loc[merged_tweets['AAPL-outliers'] == 1]
```

It's hard to tell by looking at the graph, but looking through the labels, higher numbers are consistently classified as outliers while lower nubmers are not. It's a little dissatisfying because it seems a little off to call so many numbers outliers. Let's try one more approach. 

Next we will try to use the LocalOutlierFactor classifier


```python
from sklearn.neighbors import LocalOutlierFactor

X = merged_tweets[['AAPL', 'total_volume']]
clf = LocalOutlierFactor()
classifications = pd.DataFrame(clf.fit_predict(X))
classifications = classifications.replace({1: 0, -1: 1})
merged_tweets['AAPL-outliers'] = classifications
merged_tweets[['AAPL', 'AAPL-outliers']].plot()
```

```python
merged_tweets[['AAPL', 'AAPL-outliers']].loc[merged_tweets['AAPL-outliers'] == 1]
```

This strategy certainly classified less values as outliers (93 vs 1663). Furthermore, it seemed to do a good job of classifying very low numbers as outliers, as well as very high ones. It also classified some 'normal' looking numbers as outliers, so let's investigate to see if those are outliers in the context of AAPL taking up the lion's share of tweet volume.


```python
merged_tweets[['AAPL', 'AAPL-outliers', 'total_volume']].loc[merged_tweets['AAPL-outliers'] == 1]
```

It's common for AAPL to take up the majority of volume over a given period, and we see that that is not the case for many of the outliers!
